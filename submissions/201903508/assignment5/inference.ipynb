```python
# ==========================================
# [Cell 1] ì¶”ë¡  í™˜ê²½ ì„¤ì • (Setup)
# ==========================================
import joblib
import re

# ëª¨ë¸ ë¡œë“œ í•¨ìˆ˜
def load_model():
    try:
        model = joblib.load('models/sentiment_model.joblib')
        vectorizer = joblib.load('models/tfidf_vectorizer.joblib')
        return model, vectorizer
    except FileNotFoundError:
        print("âŒ ëª¨ë¸ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤. 'training.ipynb'ë¥¼ ë¨¼ì € ì‹¤í–‰í•´ì£¼ì„¸ìš”.")
        return None, None

model, vectorizer = load_model()

# ì „ì²˜ë¦¬ í•¨ìˆ˜ (Trainingê³¼ ë™ì¼í•˜ê²Œ ìœ ì§€)
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

# ì˜ˆì¸¡ í•¨ìˆ˜
def predict_sentiment(text):
    if model is None: return
    
    # 1. ì „ì²˜ë¦¬
    cleaned = clean_text(text)
    # 2. ë²¡í„°í™”
    vec = vectorizer.transform([cleaned])
    # 3. ì˜ˆì¸¡
    pred = model.predict(vec)[0]
    prob = model.predict_proba(vec)[0]
    
    label = "Bullish ğŸš€ (ê¸ì •)" if pred == 1 else "Bearish ğŸ“‰ (ë¶€ì •)"
    confidence = prob[pred] * 100
    
    return label, confidence

print("âœ… ì¶”ë¡  ì‹œìŠ¤í…œ ì¤€ë¹„ ì™„ë£Œ")

# ==========================================
# [Cell 2] ì‹¤ì œ ì¶”ë¡  ì˜ˆì‹œ (Real-world Examples)
# ==========================================
# ê³¼ì œ ìš”êµ¬ì‚¬í•­: ìµœì†Œ 3ê°œ ì´ìƒì˜ ì‹¤ì œ ì˜ˆì‹œ í¬í•¨

examples = [
    "Nvidia earnings were amazing, stock is going to skyrocket!",  # ëª…í™•í•œ ê¸ì •
    "The inflation data looks bad, market will crash tomorrow.",   # ëª…í™•í•œ ë¶€ì •
    "I am not sure what to do with my Tesla shares."               # ì• ë§¤í•˜ê±°ë‚˜ ì¤‘ë¦½
]

print("--- ğŸ” Real Usage Inference Test ---")
for i, text in enumerate(examples):
    label, conf = predict_sentiment(text)
    print(f"\n[Case {i+1}]")
    print(f"Input: \"{text}\"")
    print(f"Result: {label}")
    print(f"Confidence: {conf:.2f}%")

# ==========================================
# [Cell 3] ì‚¬ìš©ì ì…ë ¥ í…ŒìŠ¤íŠ¸ (Interactive)
# ==========================================
# ì‚¬ìš©ìê°€ ì§ì ‘ ì…ë ¥í•´ë³¼ ìˆ˜ ìˆëŠ” ì½”ë“œ (ì„ íƒì‚¬í•­)
user_input = "Apple just announced a new revolutionary product." # ì—¬ê¸°ì— ì›í•˜ëŠ” ë¬¸ì¥ì„ ë„£ì–´ë³´ì„¸ìš”
label, conf = predict_sentiment(user_input)
print(f"\n[User Test]")
print(f"Input: \"{user_input}\"")
print(f"Result: {label} ({conf:.2f}%)")
```
