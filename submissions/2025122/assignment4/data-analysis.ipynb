```python
import json
with open("data/data.json", 'r') as fp:
    data = json.load(fp)
```

```python
from collections import Counter
label_distribution = Counter([x['relevance'] for x in data])
print(label_distribution)
```

```python
from collections import Counter
length_distribution = Counter([len(x['abstract'].split()) for x in data])

for length, frequency in length_distribution.most_common():
    print(f"# of {length}-words abstract: {frequency}")
```
